```python
# Import necessary libraries
import pandas as pd
```

## A. Grocery Market Data

```python
# Create dataframe from USDA data
usda_df = pd.read_csv('MNRetailDatasetCSV.CSV')
```

```python
# Take a look at what needs done
usda_df.head()
```

```python
# Looks like columns need standardized to snake_case
usda_df.columns = usda_df.columns.str.lower()

usda_df.head()
```

```python
# What's going on in the level_of_trade column?
usda_df['level_of_trade'].unique()

# If the only value in the column is 'RETAIL', then it seens like it can be dropped
usda_df.drop('level_of_trade', axis=1, inplace=True)

usda_df.head()
```

```python
# What's going on in the frequency column?
usda_df['frequency'].unique()

# Only 'WEEKLY'.  Can be dropped.
usda_df.drop('frequency', axis=1, inplace=True)

usda_df.head()
```

```python
# What's going on in issueing_office? 
usda_df['issuing_office'].unique()

# Huh... that's a surprisingly small number of different offices listed.  Still useful to keep, though.
```

```python
# What's going on in report_date?
usda_df['report_date'].unique()

# Gez... how many unique dates are in this column?
usda_df['report_date'].nunique()

# 252.  What's the earliest date and the latest date?
usda_df['report_date'] = pd.to_datetime(usda_df['report_date'])
earleist_date = usda_df['report_date'].min()
latest_date = usda_df['report_date'].max()

print(earleist_date)
print(latest_date)

# Looks like we have 2020-01-03 to 2024-08-30 in this dataset.  
```

```python
# What's going on in program?
usda_df['program'].unique()

# Just 3 values, eh... I was expecting more.  I guess we'll see what's in level_1, etc.
# Do we need this column??? 
```

```python
# What's going on in level_1?
usda_df['level_1'].nunique()

# 82 unique values.  What kinds of values do we have?
usda_df['level_1'].unique()
```

```python
# What's going on in level_2?
usda_df['level_2'].unique()

# How much of this column is NaN? 
nan_percentage = usda_df['level_2'].isna().sum() / len(usda_df['level_2']) * 100
print(nan_percentage)

# 8.06%.  Do we need these sub-categories on items, though? 
# Guess I'll keep it for now... 
```

```python
# What's going on in level_3?
usda_df['level_3'].unique()

# Definitely don't think this is needed
usda_df.drop('level_3', axis=1, inplace=True)

usda_df.head()
```
